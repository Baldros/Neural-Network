````python
dados = genfromtxt('data_banknote_authentication.txt', delimiter=',')
dados
````

## Tratamento inicial dos dados:

Os dados fornecidos são um imenso array... Então a
ideia aqui é separar esse array em dois, um contendo
os labels (rótulos/etiquetas) que não nossos resultados
 e features que são nossas informações. 

Esse tratamento inicial é uma coisa que todos os dados que venham nessa forma precisam ser feitos. Por que sempre, para esse tipo de trabalho, precisa-se
dos dados da informação e dos rótulos separadamente.

````python
#Puxando os rótulos
labels = dados[:,4] #Assim, labels será um array apenas com os rótulos.
labels
````



```
Também não entendi porque ",4", apenas copiei como
o professor copiou. De mesmo modo, puxer-se-irá as 
caracteristicas.
```



````python
#Puxando as caracteristicas:
features = dados[:,0:4]
features
````

````python
#Datasets
X = features #X maiúsculo é sempre o dataset
y = labels #y minúsculo é sempre os rótulos
````



```
Por fim, temos os datasets divididos e tudo certinho.
```



## Separando os dados:

Os dados precisam ser separados entre dados para treino
e dados para teste, para o melhor funcionamento das
redes neurais.

````python
#Importando a função para a divisão dos dados:
from sklearn.model_selection import train_test_split
````

````python
#Separando dados para treinamento
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state=42)
````



```
As vezes, os subconjuntos para treino e teste ficam
num escalonamento estranho. Para resolver isso, 
precisa-se passar os dados de treino e teste por um
tratamento específico.
```



## Reprocessando os dados:

Não é algo essencial para esses dados, mas é uma
boa prática, pois dados datasets maiores precisam ser
normalizados para terem menor custo computacional.

````python
#Importando a função para o reprocessamento dos dados:
from sklearn.preprocessing import MinMaxScaler
````

````python
#Reprocessando os dados:
scaler_object = MinMaxScaler()
````

````python
scaler_object.fit(X_train)
scaler_object.fit(X_test)
````

````python
scaled_X_train = scaler_object.transform(X_train)
scaled_X_test = scaler_object.transform(X_test)
````

````python
X_train.max()
````

````python
scaled_X_train.max()
````



```
A ideia desse reprocessamento dos datasets é ganho de
performance... Nesse caso específico não fará tanta
diferença, mas para datasets maiores, é bom esse
tratamento pra deixar os dados bem comportados.

Ou seja, todo esse processo de scaler_object, fit e
transform nada mais é do que uma boa prática de normalização de dados.
```



## Criando o modelo com Keras:

````python
#Importação das bibliotecas:
from keras.models import Sequential #Cria a rede neural de forma sequenciada
from keras.layers import Dense #Primeira camada de neurónios
````

### Criação do modelo:

````python
#Criando as variaveis Modelo:
modelo = Sequential() #Criando o modelo Neural

modelo.add(Dense(4, input_dim=4,activation='relu')) #Primeira camada de neuronios
modelo.add(Dense(8,activation='relu')) #Segunda camada de neuronios
modelo.add(Dense(1,activation='sigmoid')) #Teceisa cama de neuronios - Sigmoide pq a saida é 0 ou 1
````

````python
#Função de custo:
modelo.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
````

### Treinamento do Modelo:

````python
#Treinando os modelos:
modelo.fit(scaled_X_train, y_train, epochs=50, verbose=2)
````




````python
#Prevendo:
modelo.predict(scaled_X_test)
````



```

```



````python
#Checando acuracidade:
predictions = (modelo.predict(scaled_X_test) > 0.5).astype("int32")
predictions
````

````python
#Checando métricas:
modelo.metrics_names
````

````python
#Checando a performance do modelo:
modelo.evaluate(x=scaled_X_test,y=y_test)
````
